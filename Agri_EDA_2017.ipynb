```python
import pandas as pd 
import numpy as np 
```

```python
data_a = pd.read_csv("Resources/Cierre_agricola_mun_2017.csv", encoding='latin-1')
data_a.head()
```

```python
# Check Dataset Size
data_a.shape
```

```python
# Check datatypes
data_a.dtypes
```

```python
# Check for nulls
data_a.isnull().sum()
```

```python
# Check columns names & index
for i, v in enumerate(data_a):
    print(i,v)
```

```python
# Drop Columns
data_a = data_a.drop(data_a.columns[[3,4,5, 6, 9, 11, 13, 15, 24]], axis=1)
```

```python
# Display Data
data_a
```

```python
# Calculate percentage of cosecha vs sembrada
data_a['Sembradavscosecha'] =  data_a['Cosechada'] / data_a['Sembrada']
data_a 
```

```python
# Profit by Hectare
data_a['Valorxhectarea'] = data_a['Valorproduccion'] / data_a['Cosechada']
data_a
```

```python
# Fill NAN values with zero
data_a = data_a.fillna(0)
```

```python
# Rename columns
data_a.rename(columns = {"Idestado":"Idest", "Idmunicipio":"Idmuni","Nomestado": "Estado", 
                         "Nommunicipio": "Municipio", "Nomcicloproductivo":"Cicloproductivo", 
                         "Nommodalidad": "Modalidad","Nomunidad":"Unidad", 
                         "Nomcultivo":"Cultivo"}, inplace=True)
```

```python
# Convert to lower case the column names
data_a.columns = data_a.columns.str.lower()
```

```python
data_a
```

```python
#Import center coordinates for each Municipio
data_coords = pd.read_csv("Resources/coordenadas.csv", encoding="UTF-8")
data_coords.head(10)
```

```python
data_coords_mun = data_coords[data_coords["CVE_LOC"] == 1]
data_coords_mun
```

```python
# Check datatypes
data_coords.dtypes
```

```python
# Drop Columns
data_coords_mun = data_coords_mun.drop(data_coords_mun.columns[[0,2,3,5,6,7,11]], axis=1)
```

```python
data_coords_mun.head(10)
```

```python
data_coords_mun.groupby(['NOM_ENT','NOM_MUN'])['LAT_DEC','LON_DEC'].mean().reset_index()
```

```python
data_coords_mun
```

```python
data_coords_mun.NOM_ENT[data_coords_mun.NOM_ENT=="Michoacán de Ocampo"] = "Michoacán"
data_coords_mun.NOM_ENT[data_coords_mun.NOM_ENT=="Coahuila de Zaragoza"] = "Coahuila"
data_coords_mun.NOM_ENT[data_coords_mun.NOM_ENT=="Veracruz de Ignacio de la Llave"] = "Veracruz"
data_coords_mun
```

```python
# Rename columns
data_coords_mun.rename(columns = {"NOM_ENT":"estado", "NOM_MUN":"municipio","LAT_DEC": "latitud", 
                         "LON_DEC": "longitud", "ALTITUD":"altitud"}, inplace=True)
```

```python
data_coords_mun
```

```python
data_a
```

```python
# Merge
data_complete = pd.merge(data_a, data_coords_mun, on=['estado', 'municipio'], how='left') 
data_complete
```

```python
data_complete.to_csv("Resources/clean_dataset_2017.csv",encoding='latin-1', index=False)
```
